```python
 def hh_parse (link,user_agent):
    jobs=[]
    urls=[]
    
    urls.append(link)
    session = requests.Session()
    request = session.get(link,headers = user_agent)
    if request.status_code == 200:
        soup = bs(request.content,'html.parser')
        try:
            pagitation = soup.find_all('a',{'data-qa':'pager-page'})
            count =int(pagitation[-1].text)
            for i in range(count):
                url = f'https://ufa.hh.ru/search/vacancy?L_is_autosearch=false&area=1202&clusters=true&enable_snippets=true&text=python&page={i}'
                if url not in urls:
                    urls.append(url)          
        except:
            pass
    for url in urls:
        request = session.get(url,headers=user_agent)
        soup = bs(request.content,'lxml')
        divs = soup.find_all('div',{'class':'vacancy-serp-item'})
        for div in divs:
            try:
                title = div.find("a", attrs={"data-qa": "vacancy-serp__vacancy-title"}).text
                href = div.find("a", attrs={"data-qa": "vacancy-serp__vacancy-title"})["href"]
                price = div.find('div',{'class':'vacancy-serp-item__compensation'}).getText()
                p1=price.split('\xa0')
                p2=''.join(p1)
                jobs.append({
                    'vacancy':title,
                    'site':href,
                    'price':p2
                })
            except:
                pass
    else:
        print(1)
    
    hhru.append(jobs)       
hh_parse(link,user_agent)
```

```python
def sj_parse (list_url,user_agent):
    jobs_sj =[]
    for i in list_url:
        session = requests.Session()    
        request = session.get(i,headers = user_agent)
        if request.status_code == 200:
            soup = bs(request.content,'html.parser')
            divs = soup.find_all('a',{'target':'_blank'})
            price = soup.find_all('span',{'class':'_3mfro _2Wp8I f-test-text-company-item-salary PlM3e _2JVkc _2VHxz'})
            for p,i in zip(price,divs):
                val = p.text
                p1 = val.replace('\xa0','').replace('—', ' ').replace('₽',' ')    
                p2 = ''.join(p1)
                target = i.find('div',{'class':'_3mfro CuJz5 PlM3e _2JVkc _3LJqf'})
                vac = target.text
                link = i.get('href')
                jobs_sj.append({
                'vacancy':vac,                    
                'site':f'superjob.ru{link}',
                'price':p2
                })
        pro.append(jobs_sj)
        
sj_parse(list_url,user_agent)
```
